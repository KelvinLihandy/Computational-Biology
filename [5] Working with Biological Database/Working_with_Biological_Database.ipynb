```python
from Bio import SeqIO
from Bio.PDB import PDBParser
from Bio.PDB import MMCIFParser
from Bio.SeqUtils import GC, MeltingTemp
import matplotlib.pyplot as plt
```

```python
fasta = SeqIO.read("sequence.fasta", "fasta")
print(fasta)
print(fasta.seq)
```

```python
gb = SeqIO.read("sequence.gb", "genbank")
print(gb)
print(gb.seq)
```

```python
pdbParser = PDBParser()
structure = pdbParser.get_structure("4CIW", "4ciw.pdb")
model = structure[0]
for c in model:
    print("chain", c)
    for r in c:
        print("|-> res", r)
        for a in r:
            print("    |-> atom",a)
```

```python
mmcif = MMCIFParser()
structure = mmcif.get_structure("9GLT", "9glt.cif")
model = structure[0]
for c in model:
    print("chain", c)
    for r in c:
        print("|-> res", r)
        for a in r:
            print("    |-> atom",a)
```

```python
seqY = fasta.seq
length = round(len(seqY)/2)
seqA = seqY[:length]
seqB = seqY[length:]

gcA = GC(seqA)
gcB = GC(seqB)
mtA = MeltingTemp.Tm_NN(seqA)
mtB = MeltingTemp.Tm_NN(seqB)
_, axs = plt.subplots(1, 2)
axs[0].bar(["seqA", "seqB"], [gcA, gcB])
axs[1].bar(["seqA", "seqB"], [mtA, mtB])
plt.show()
```
